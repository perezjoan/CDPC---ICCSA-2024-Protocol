```python
# EMC2 WP 2 Processing 1.0 (S : subset)
# Morphometry on buildings : handle missing values in building height and number of floors ; generate new indicators for building data : surface area,
# perimeter, elongation, convexity, floor area, and various products of these indicators ; morphological tessellation of buildings and calculates Queen
# contiguity spatial weights of order 3 and 6 and median values of aformentionned indicators.
# Author : Perez Joan
# Input data : output from EMC2 WP 2 Processing 0.1 (department or subset)
```

```python
# 0.1  Packages, local filepaths & parameters

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import geopandas as gpd
import numpy as np
import momepy
import libpysal

# Specify the location of subset
gpkg_s = "C:\\Users\\jperez\\Documents\\Current 1\\emc2\\Output\\WP2_DPC_0.1_Subset.gpkg"

# Specify the file path to record the results on the Subset
gpkg_save_s = "C:\\Users\\jperez\\Documents\\Current 1\\emc2\\Output\\WP2_DPC_1.1_Subset.gpkg"
```

```python
# 1  Complete NA values for building height (hauteur) and number of floors (nombre_d_etages)
# Read the data
building_subset = gpd.read_file(gpkg_s, layer="building")

# fill missing height (hauteur)
building_subset['hauteur'] = building_subset['hauteur'].fillna(building_subset['altitude_minimale_toit'] - building_subset['altitude_minimale_sol'])
# checks if the 'hauteur' column is NaN and if the 'nombre_d_etages' is not NaN.
# If both conditions are met : multiplies the value of 'nombre_d_etages' by 3 and assigns it to 'hauteur'
building_subset['hauteur'] = building_subset.apply(lambda row: row['nombre_d_etages'] * 3 if pd.isna(row['hauteur'])
                                     and not pd.isna(row['nombre_d_etages']) else row['hauteur'], axis=1)
# checks if the 'hauteur' column is NaN and if the 'nombre_d_etages' is not NaN.
# If both conditions are met : multiplies the value of 'nombre_d_etages' by 3 and assigns it to 'hauteur'
building_subset['nombre_d_etages'] = building_subset.apply(lambda row: round(row['hauteur'] / 3) if pd.isna(row['nombre_d_etages'])
                                             and not pd.isna(row['hauteur']) else row['nombre_d_etages'], axis=1)
print('Number of remaining values with NA for height/floors : ', building_subset[['hauteur', 'nombre_d_etages']].isna().all(axis=1).sum())
```

```python
# 2  Basic new indicators : Surface, Compacity, Elongation, Product [1-E].C.S, Product [1-E].S

# ID
building["ID"] = range(len(building))

# Surface
building['area'] = building.geometry.area

# Calculating perimeter
building['perimeter'] = building.geometry.length

# Calculating elongation
building['elongation'] = momepy.Elongation(building).series

# Convexity
building['convexity'] = momepy.Convexity(building).series

# Floor area
building['FA'] = building['nombre_d_etages'] * building['area']

# Product [1-E].C.S
building['ECS'] = (1 - building['elongation']) * building['area'] * building['convexity']

# [1-E].S
building['ES'] = (1 - building['elongation']) * building['area']

# Shared walls
building["SW"] = momepy.SharedWallsRatio(building).series
```

```python
# 2  Adjency matrix + contiguity of order 3 and 6

# morphological tessellation (pseudo-plot)
limit = momepy.buffered_limit(building, 100)
tessellation = momepy.Tessellation(building, "ID", limit, verbose=False, segment=1)
tessellation = tessellation.tessellation

# Queen contiguity calculation
queen_1 = libpysal.weights.contiguity.Queen.from_dataframe(tessellation, ids="ID", silence_warnings=True)

# Generate spatial weights based on Queen contiguity of order k = 3 (all features within 3 spatial steps are adjacent)
queen_3 = momepy.sw_high(k=3, weights=queen_1)

# Generate spatial weights based on Queen contiguity of order k = 6 (all features within 6 spatial steps are adjacent)
queen_6 = momepy.sw_high(k=6, weights=queen_1)
```

```python
# 3  Indicators at adjency matrix and queen contiguity of order 3 and 6 level

# Floor-area
median = building['FA'].median()
building['median_FA_ADJ1'] = momepy.AverageCharacter(building, values='FA', spatial_weights=queen_1, unique_id='ID').median
building['median_FA_ADJ3'] = momepy.AverageCharacter(building, values='FA', spatial_weights=queen_3, unique_id='ID').median
building['median_FA_ADJ3Gap'] = building['median_FA_ADJ3'] - median
building['median_FA_ADJ6'] = momepy.AverageCharacter(building, values='FA', spatial_weights=queen_6, unique_id='ID').median
building['median_FA_ADJ6Gap'] = building['median_FA_ADJ6'] - median

# ECS
median = building['ECS'].median()
building['median_ECS_ADJ1'] = momepy.AverageCharacter(building, values='ECS', spatial_weights=queen_1, unique_id='ID').median
building['median_ECS_ADJ3'] = momepy.AverageCharacter(building, values='ECS', spatial_weights=queen_3, unique_id='ID').median
building['median_ECS_ADJ3Gap'] = building['median_ECS_ADJ3'] - median
building['median_ECS_ADJ6'] = momepy.AverageCharacter(building, values='ECS', spatial_weights=queen_6, unique_id='ID').median
building['median_ECS_ADJ6Gap'] = building['median_ECS_ADJ6'] - median

# CS
median = building['ES'].median()
building['median_ES_ADJ1'] = momepy.AverageCharacter(building, values='ES', spatial_weights=queen_1, unique_id='ID').median
building['median_ES_ADJ3'] = momepy.AverageCharacter(building, values='ES', spatial_weights=queen_3, unique_id='ID').median
building['median_ES_ADJ3Gap'] = building['median_ES_ADJ3'] - median
building['median_ES_ADJ6'] = momepy.AverageCharacter(building, values='ES', spatial_weights=queen_6, unique_id='ID').median
building['median_ES_ADJ6Gap'] = building['median_ES_ADJ6'] - median

# SW
median = building['SW'].median()
building['median_SW_ADJ1'] = momepy.AverageCharacter(building, values='SW', spatial_weights=queen_1, unique_id='ID').median
building['median_SW_ADJ3'] = momepy.AverageCharacter(building, values='SW', spatial_weights=queen_3, unique_id='ID').median
building['median_SW_ADJ3Gap'] = building['median_SW_ADJ3'] - median
building['median_SW_ADJ6'] = momepy.AverageCharacter(building, values='SW', spatial_weights=queen_6, unique_id='ID').median
building['median_SW_ADJ6Gap'] = building['median_SW_ADJ6'] - median
```

```python
# 4 Write a layer of buildings with morphology indicators
building.to_file(gpkg_save_s, layer="Buildings_with_morphology_indicators", driver="GPKG")
```
